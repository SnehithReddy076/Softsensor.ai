```python
from bs4 import BeautifulSoup 
import requests 

headers = { 
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36 OPR/89.0.4447.51" #
}

url="https://www.pathologyoutlines.com" 
r1 = requests.get(url, headers=headers).text 
```

```python
organ_content={}
```

```python
soup1 = BeautifulSoup(r1, 'lxml') 
columns=soup1.find_all('div', class_='home_content clearfix')
 
```

```python
urls=[] 
organs=[] 
for column in columns: 
    links=column.findAll('a') 
    for a in links: 
        urls.append(a['href']) 
    for organ in column.text.strip().split('\n'): 
        if organ:
            organs.append(organ) 
```

```python
print(organs)
```

```python
for num in range(len(urls)): 
    content=[] 
    url=urls[num] 
    r2 = requests.get(url, headers=headers).text 
    soup2=BeautifulSoup(r2,'lxml') 
    subsecs=soup2.find_all('span',class_='toc_subsection toc_links') 
    for subsec in subsecs:
        for sublink in subsec.text.split('\n'): 
            if sublink: 
                content.append(sublink)
    finalsec=soup2.find('div',class_='toc_subsection toc_links') 
    if finalsec: 
        finalseclist=finalsec.text.split('\n')
        finalseclist=[x for x in finalseclist if x] 
        content=content+finalseclist[1:]
    organ_content[organs[num]]=[] 
    if not content:
        subsecs=soup2.find_all('div',class_='toc_links') 
        content=subsecs[0].text.rstrip().split('\n') 
        content=[x for x in content if x] 
        content=[x for x in content if not x.endswith(':')] 
    for c in content: 
        if c: 
            organ_content[organs[num]].append(c) 
```

```python
print("Name of organs for which dictionary is created:\n-----------------------------------------------------")
for keyw in organ_content:
    print(keyw)
```

```python
for key in organ_content.keys():
    print(key,end=": \n") 
    for content in organ_content[key]: 
        print(content,end=", ") 
    print('\n') 
```
